```python
category_names = ['Základní', 'Vyučen', 'Maturita', 'VŠ']
results = {
    'Vzdělání': [data['Vzdělání'].value_counts()['Základní'], 
                 data['Vzdělání'].value_counts()['Vyučen'], 
                 data['Vzdělání'].value_counts()['Maturita'], 
                 data['Vzdělání'].value_counts()['VŠ']]
}


def survey(results, category_names):
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.colormaps['RdYlGn'](np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, len(results)))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.5, label=colname, color=color)

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        ax.bar_label(rects, label_type='center', color=text_color)
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 1), loc='lower left', fontsize='small')
    return fig, ax


survey(results, category_names)
plt.show()
```

```python
def merge_vysledek(column):
    arr = column.value_counts().values
    soucet = arr.sum()
    return arr
def vyhodnot(input, zajem_stupne):
    vysledek = {}
    for column in input.columns:
        vysledek[column] = np.array(merge_vysledek(input[column])) 
    return vysledek

zajem_stupne = ['Velmi mně zajímá', 'Částečně /někdy mě zajímá', 'Málo mě zajímá', 'Vůbec mě nezajímá']
temata = data.columns[13:21]

vysledky = vyhodnot(data[temata], zajem_stupne)
survey(vysledky, zajem_stupne)
plt.show()
```

```python
def zpracuj_odpoved(data, od, do, dropped_labels):
    result = data[data.columns[od]].value_counts()
    
    for i in range(do - od - 1):
        result = result.add(data[data.columns[od + i + 1]].value_counts(), fill_value=0)
        
    result = result.sort_values(ascending=False)
    result = result.drop(labels=dropped_labels)
    
    return result

duvod = zpracuj_odpoved(data, 31, 33, [-9, 'Jiné', 'Neuvedl'])
plt.pie(duvod.values, labels = duvod.index, autopct='%i')
plt.show()
```
